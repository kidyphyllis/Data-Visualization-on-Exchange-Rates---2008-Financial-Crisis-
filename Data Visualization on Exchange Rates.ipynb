# Data Visualization on Exchange Rates

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. Daria Chemkaeva put together the data set and made it available on Kaggle — the data source is the European Central Bank.

```python
import pandas as pd
exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")
exchange_rates.info()
```

 ## Data Cleaning
 The project will be focused on the exchange rate between the Euro and the American dollar. 

```python
# Rename the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar', 'Period\\Unit:': 'Time'}, inplace=True)

#Change the Time column to a datetime data type.
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

#Sort the values by Time in ascending order.
exchange_rates.sort_values('Time', inplace=True)

#Reset the index (and drop the initial index).
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
# Isolate the Time and the US_dollar columns. Assign them to euro_to_dollar.
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()

# Drop all the 62 rows where the - character appears in the US_dollar column. 
```

```python
# Convert the US_dollar column to a float data type.
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

euro_to_dollar.info()
```

## Plot
Generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#Display graphs inline

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
# Rolling Mean to include seven (one week), 30 (one month), 60 or 90 days, or 365 days (one year).
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 60, 90, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')

# Auto-adjusts the padding between subplots    
plt.tight_layout() 
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## Euro - Dollar Rate during the 2007 - 2008's Financial Crisis Visualizations
We need to investigate how the euro-dollar rate changed during the 2007-2008's financial crisis. We can also show the data for 2006 and 2010 for comparison. We can use a line plot.

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar.Time.dt.year >= 2007) & (euro_to_dollar.Time.dt.year <= 2008)]
```

```python
### Plot of financial crisis by the rolling mean
fig,ax = plt.subplots(figsize=(12,4))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'], linewidth=2, color='#B1D780')

### Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'], financial_crisis_7_8['rolling_mean'], linewidth=3, color='#a23d28')

### Highlihting the peak of the crisis
ax.axvspan(xmin=733112.0, xmax=733302.0, ymin=0.09,  alpha=0.3, color='grey')
    
### Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 732272.0
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=12)
    x += 365

y = 1.193   
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(732172.0, y, rate, alpha=0.5, fontsize=12)
    y += 0.1

## Adding a title and a subtitle    
ax.text(732172.0, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold')
ax.text(732172.0, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)

### Adding a signature
ax.text(732172.0, 1.07, 'Graph By Phyllis Wanjiku Gitu' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#1d1d1d',
        size=12)

### Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```
